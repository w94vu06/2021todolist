```python
!pip install yfinance
#載入需要的文件

### 科學計算，數據處理常用的基礎套件
import pandas as pd

#作圖，視覺化套件
import matplotlib.pyplot as plt
import plotly.graph_objects as go 
import matplotlib as mpl

# yahoo finance 套件，用來下載股價資訊
import yfinance as yf
plt.style.use('fivethirtyeight')
```

```python
#開始
###台股: 2330.TW ，3008.TW ，2317.TW
stock_no='2330.TW'
#起始日期
start_date = '2021-05-01'

#下載資料會自動抓到最近的交易日期
df = yf.download(stock_no, start=start_date)
```

```python
#前五筆
df.head()
```

```python
#後五筆
df.tail(5)
```

```python
#SHOW圖
figure = go.Figure(data= [go.Candlestick(x=df.index,
open=df['Open'],
high=df['High'],
low=df['Low'],
close=df['Close'],
increasing_line_color='red',
decreasing_line_color='green')])

figure.update_layout(title= '2330 Price 台積電',xaxis_title = 'Date 日期',yaxis_title= 'Price 價格')

#SHOW圖
figure.show()
```
